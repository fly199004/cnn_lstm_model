```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# 加载数据并提取 'features'
data = np.load('bert_vectors/batch_0000.npy', allow_pickle=True).item()
word_vectors = data['features'].reshape(-1, data['features'].shape[-1])  # 重新形状为 (4096, 768)

print("Shape of word vectors:", word_vectors.shape)

# 使用t-SNE进行降维
tsne = TSNE(n_components=2, random_state=0, perplexity=min(30, len(word_vectors)-1))
word_vectors_2d = tsne.fit_transform(word_vectors)

# 绘图
plt.figure(figsize=(16, 9))
plt.scatter(word_vectors_2d[:, 0], word_vectors_2d[:, 1], alpha=0.5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Spatial visualization of word vectors')
plt.show()
```
